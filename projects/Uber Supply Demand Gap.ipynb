# Problem:
You all must have travelled to and from the airport. Have you anytime used Uber or in fact any other cab service to travel to and from the airport? Did you at any time face the problem of cancellation by the driver or the issue of non-availability of cars? Well if these are the problems faced by customers, these very issues also impact the business of Uber. If drivers cancel the request of riders or if cars are unavailable. Put yourself into the shoes of Uber analyst and analyse the available data to understand why such issues might be occuring.

In the data, as you can see, there are 6 attributes associated with each request id. These are:-
1. Date of request: The date on which the customer made the trip request
2. Time of request: The time at which the customer made the trip request
3. Drop-off time: The drop-off time, in case the trip was completed 
4. Pick-up point: The point from which the request was made
5. Driver id: The unique identification number of the driver
6. Status of the request: The final status of the trip, that can be either completed, cancelled by the driver or no cars available

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plot
import seaborn as sns
%matplotlib inline
```

```python
uber_data = pd.read_csv('Uber.csv')
uber_data.head()
```

### aggregating the data by hour

```python
uber_data['hour'] = uber_data['Request time'].apply(lambda x: x[0:x.index(':')]).astype(int)
```

```python
#plotting the number of trips by hour
plot.figure(figsize=(10, 7))
sns.countplot(x='hour', hue='Pickup point',data=uber_data)
```

```python
#making slots by number of trips
conditions = [(uber_data.hour < 4),
              (uber_data.hour < 10),
              (uber_data.hour < 15),
              (uber_data.hour < 22),
              (uber_data.hour >= 22)]
choices = ['Pre_Morning',"Morning_rush","day_time","Evening_rush","late_night"]
uber_data['time_slot'] = np.select(conditions, choices)
```

```python
plot.figure(figsize=(10, 7))
sns.countplot(x='time_slot', hue='Pickup point',data=uber_data)
```

## Count of Different time slots

```python
uber_data['time_slot'].value_counts()
```

## Count by Different pickups

```python
uber_data.groupby(['Pickup point','time_slot']).size()
```

```python
# plotting and identifying the most critical problems before Uber
plot.figure(figsize=(10, 7))
sns.countplot(x='time_slot', hue='Status',data=uber_data)
plot.xlabel('Time Slot')
plot.ylabel('Trip Response')
plot.title('Request details based on time slot and status')
```

```python
#problem 1
morning_rush_data = uber_data.loc[uber_data['time_slot']=="Morning_rush",]
```

```python
plot.figure(figsize=(10, 7))
sns.countplot(x='Pickup point', hue='Status',data=morning_rush_data)
plot.xlabel('Picup points')
plot.ylabel('Trip Response')
plot.title('Trip Response by Picup point and status in morning hours')
```

```python
#problem 2
evening_rush_data = uber_data.loc[uber_data['time_slot']=="Evening_rush",]
plot.figure(figsize=(10, 7))
sns.countplot(x='Pickup point', hue='Status',data=evening_rush_data)
plot.xlabel('Picup points')
plot.ylabel('Trip Response')
plot.title('Trip Response by Pickup point and status in evening hours')
```

Conclusion:
* There are more flights are coming to city in evening time, Hence the supply in the evening hour is less at airport then demand leading to unavailability of cab and cancellation
* There are more taking off in the morning time, Hence the supply in the morning hour is less at in city then demand leading to unavailability of cab and cancellation
